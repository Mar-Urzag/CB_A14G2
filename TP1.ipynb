```python
#3) Codificar un programa en Python que permita ingresar los coeficientes de una funci贸n lineal y 
#de una funci贸n cuadr谩tica, y como resultado arroje las coordenadas de los puntos de intersecci贸n 
#de ambas curvas, en el caso de ser valores reales. El programa adem谩s deber谩 informar al usuario 
#si las funciones son secantes, tangentes o exteriores una con otra.  (40 puntos) 
#Ejemplo: 
 	       
#Puntos en com煤n P1(1 ; 3)   P2(2; 4)   隆Las funciones son secantes! 

# ingresar coeficientes de funcion lineal
# y = m x + p
def ingresar_coeficientes_funcion_lineal():
    m = float(input("Ingrese la pendiente de la recta: "))
    p = float(input("Ingrese la ordenada al origen de la funcion lineal: "))
    return m, p
    
# ingresar coeficientes de funcion cuadratica
# y = a x^2 + b x + c
def ingresar_coeficientes_funcion_cuadratica():
    a = float(input("Ingrese el coeficiente de la funcion cuadratica: "))
    b = float(input("Ingrese el coeficiente de la funcion cuadratica: "))
    c = float(input("Ingrese el coeficiente de la funcion cuadratica: "))
    return a, b, c

# calcular el punto de interseccion de la funcion lineal y de la funcion cuadratica
def calcular_punto_interseccion(m, p, a, b, c):
    try:
        x = (-b + (b**2 - 4*a*c)**0.5)/(2*a)
        y = m*x + p
        return x, y
    except ValueError:
        print("No se puede calcular el punto de interseccion")
        return None 

# graficar las funciones y destacar los puntos de interseccion en el grafico
def graficar_funciones(m, p, a, b, c):
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.linspace(-5, 5, 100)
    y = m*x + p
    plt.plot(x, y, label="Funcion lineal")
    y = a*x**2 + b*x + c
    plt.plot(x, y, label="Funcion cuadratica")
    plt.legend()
    plt.show()


# cartel de bienvenida
print("Bienvenido al programa de interseccion de funciones")

# ingresar los coeficientes de las funciones
m, p = ingresar_coeficientes_funcion_lineal()
a, b, c = ingresar_coeficientes_funcion_cuadratica()

# calcular el punto de interseccion
x, y = calcular_punto_interseccion(m, p, a, b, c)

# si el punto de interseccion no es none graficar las funciones y destacar el punto de interseccion
import matplotlib.pyplot as plt
#import numpy as np
if x != None:
    graficar_funciones(m, p, a, b, c)
    plt.plot(x, y, "ro")
    plt.show()
    print("Las funciones son secantes")
else:
    graficar_funciones(m, p, a, b, c)
    plt.show()
    print("Las funciones no son secantes")
```
